```python
#Importar librerías
import pandas as pd
import random
```

```python
#Conjunto de datos base para Instagram influencers

datos = {
    "influencer": [
        "CancunLife",
        "TravelAna",
        "BeachVibes",
        "LuxuryParadise",
        "VisitCancun",
        "TropicalMood"
    ],

    "likes": [1200, 5000, 15000, 30000, 100000, 75000],

    "comentarios": [50, 120, 300, 500, 1200, 800],

    "fecha": [
        "2025-01-10",
        "2025-02-15",
        "2025-03-08",
        "2025-04-20",
        "2025-05-01",
        "2025-05-22"
    ],

    "tipo_post": [
        "foto",
        "reel",
        "historia",
        "video",
        "reel",
        "foto"
    ]
}
```

```python
dfDatos = pd.DataFrame(datos)

dfDatos
```

```python
cantidad = 1000000
```

```python
dfInstagram = pd.DataFrame({

    "influencer": [
        random.choice(dfDatos["influencer"])
        for _ in range(cantidad)
    ],

    "likes": [
        random.randint(500, 500000)
        for _ in range(cantidad)
    ],

    "comentarios": [
        random.randint(10, 15000)
        for _ in range(cantidad)
    ],

    "fecha": [
        random.choice(dfDatos["fecha"])
        for _ in range(cantidad)
    ],

    "tipo_post": [
        random.choice(dfDatos["tipo_post"])
        for _ in range(cantidad)
    ]
})
```

```python
#insights
dfInstagram["engagement"] = (
    dfInstagram["likes"] +
    dfInstagram["comentarios"]
)
```

```python
dfInstagram.groupby("influencer")["engagement"].mean()
```
